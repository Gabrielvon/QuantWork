```python
import pandas as pd
import numpy as np
from tqdm import tqdm_notebook as tqdm

import os, sys, warnings
sys.path.append('/root/notebook/Workspace/Options/pyoption')
import RVolatility as rvol
import VannaVolga as vvimp
import SVI2 as svi
import option_numba as optnb
import option_helper as opth

warnings.filterwarnings("error")
```

```python
def svi_calibrate_iv(df):
    tmp_df = df.reset_index().assign(sviiv=np.nan)
    tmp_grp = tmp_df.groupby(['cp_flag', 'expire_date'])
    for grpn, fgrp in tmp_grp:
        _logk, _bsiv, _tau = fgrp[['logstrike', 'bsiv', 'tau']].dropna().values.T
        paras = svi.svi_raw_fit_epoch(_logk, _bsiv, _tau, epoch=1, min_valid=1, max_iter=10, tol=1e-6, verbose=False)
        _total_variance = svi.svi_raw(fgrp['logstrike'].values, paras)
        tmp_df.loc[fgrp.index, 'sviiv'] = np.sqrt(_total_variance / _tau)
    
    return tmp_df.set_index('timestamp')

def vv_calibrate_iv(df, min_strikes=5):
    tmp_df = df.reset_index()
    tmp_grp = tmp_df.groupby('expire_date')
    # vviv_list = [(grpn, k, vvimp.VannaVolgaImpliedVol(k, *vvimp.get_anchor_points(fgrp))) for grpn, fgrp in tmp_grp for k in fgrp['exercise_price'].values]
    vviv_list = []
    for grpn, fgrp in tmp_grp:
        if fgrp['exercise_price'].nunique() >= min_strikes:
            for k in fgrp['exercise_price'].values:
                try:
                    vviv = vvimp.VannaVolgaImpliedVol(k, *vvimp.get_anchor_points(fgrp))
                    vviv = vvimp.VannaVolgaImpliedVol(k, *vvimp.get_calib_anchor_points(fgrp))
                except ValueError as e:
                    vviv = np.nan
                vviv_list.append((grpn, k, vviv))
        else:
            for k in fgrp['exercise_price'].values:
                vviv_list.append((grpn, k, np.nan))
    vviv_df = pd.DataFrame(vviv_list, columns=['expire_date', 'exercise_price', 'vviv'])
    tmp_df = tmp_df.merge(vviv_df, on=['expire_date', 'exercise_price'], how='left')
    return tmp_df.set_index('timestamp')
```

```python
rawopt = pd.read_csv('/root/notebook/Repos/GH_Public/Option/OptionData.csv')
rawopt['timestamp'] = pd.to_datetime(rawopt['timestamp'])
optinfo = rawopt.copy()
```

```python
optinfo['market_price'] = optinfo[['ap1', 'bp1']].mean(1)
optinfo['exercise_price'] = optinfo['exercise_price'] * optinfo['volume_multiple'] / 10000
optinfo['expire_date'] = pd.to_datetime(optinfo['expire_date'], format='%Y%m%d')
optinfo['tau'] = (optinfo['expire_date'] - optinfo['timestamp']) / np.timedelta64(1, 'm') / (365 * 24 * 60)
optinfo['rf'] = 0.03
optinfo['q'] = 0
optinfo['logstrike'] = svi.logstrike(*optinfo[['spot_price', 'exercise_price', 'tau', 'rf', 'q']].values.T)
optinfo['bsiv'] = optnb.implied_volatility_series(*optinfo[['cp_flag', 'spot_price', 'exercise_price', 'rf', 'tau', 'q', 'market_price']].values.T, upper=5, lower=1e-6, tol=1e-6, maxiter=100)
optinfo['delta'] = optnb.delta(*optinfo[['cp_flag', 'spot_price', 'exercise_price', 'rf', 'tau', 'bsiv']].values.T)
# optinfo = optinfo.loc[(optinfo['bsiv'] < uvol) & (optinfo['bsiv'] > lvol)].copy()    # clean IV outlier
# optinfo = optinfo[optinfo['market_price'] > lprice]    # clean price outlier
optinfo['vega'] = optnb.vega(*optinfo[['spot_price', 'exercise_price', 'rf', 'tau', 'bsiv']].values.T)
```

```python
lvol = 0.05               # lower volatility acceptance limit
uvol = 3.00               # upper volatility acceptance limit
lprice = 0.0005               # lower price acceptance limit
# lvolume = 3000
```

```python
impvol_list = []
grp = optinfo.groupby('timestamp')
for gn, gp in tqdm(grp, total=grp.ngroups):
    gp1 = gp.copy()

    # svi_iv
    gp1 = gp1.loc[(gp1['bsiv'] < uvol) & (gp1['bsiv'] > lvol)].copy()    # clean IV outlier
    gp1 = gp1[gp1['market_price'] > lprice]    # clean price outlier
    gp1 = svi_calibrate_iv(gp1)
    gp1 = gp1.reset_index().set_index('index')

    # vviv
    try:
        gp1 = vv_calibrate_iv(gp1)
    except Exception as e:
        print(gn, e)
        gp1['vviv'] = np.nan

    bsiv, sviiv, vviv, sum_volume, vega = gp1[['bsiv', 'sviiv', 'vviv', 'sum_volume', 'vega']].values.T
    _w_volume_bsiv = np.nansum(bsiv * sum_volume / np.sum(sum_volume))
    _w_vega_bsiv = np.nansum(bsiv * vega / np.sum(vega))
    _w_volume_sviiv = np.nansum(sviiv * sum_volume / np.sum(sum_volume))
    _w_vega_sviiv = np.nansum(sviiv * vega / np.sum(vega))
    _w_volume_vviv = np.nansum(vviv * sum_volume / np.sum(sum_volume))
    _w_vega_vviv = np.nansum(vviv * vega / np.sum(vega))

    try:
        _vix = rvol.calc_vix(gp) / 100
    except Exception:
        _vix = np.nan

    impvol_list.append([gn, _w_volume_bsiv, _w_vega_bsiv, _w_volume_sviiv, _w_vega_sviiv, _w_volume_vviv, _w_vega_vviv, _vix])
```

```python
impvol_df = pd.DataFrame(impvol_list, columns=['timestamp', 'wvolume', 'wvega', 'cwvolume', 'cwvega', 'vwvolume', 'vwvega', 'vix']).set_index('timestamp')
```
